# Limpieza simple de Datos


Lo primero que se quiere hacer es enviar todas las columnas que se refieren al género(*'Action', 'Adventure', 'Comedy', 'Crime', 'Family', 'Fantasy','Mystery', 'Sci-Fi', 'Thriller'*) a una sola columna que diga género y contenga todos estos valores como texto


## Eliminar Columnas de Género

- Primero se llaman las librerías a usar

```python
import numpy as np
import pandas as pd
```

- Ahora se procede a llamar la base de datos que se va a manejar a lo largo del notebook

```python
df_games=pd.read_csv("./imdb-videogames.csv", index_col=0)
df_games
```

- Se llama el método *info()* para ver información general del archivo

```python
df_games.info()
```

- Ahora, para condensar los datos de género en una sola columna, empezamos por hacer una copia segmentada del dataframe para un mejor manejo de los datos

```python
change=df_games.iloc[:,-9:].copy()
change
```

- Creamos una función para remplazar los valores verdaderos por el nombre de la columna y los falson por un "-"

```python
def proof(x):
    return x.apply(lambda y: x.name if y else "-")
```

- Aplicamos la función al código

```python
change=change.apply(lambda x: proof(x))
change
```

- Se procede a crear una función que primero nos ponga todos los elementos de una fila en una lista *(1)*, una vez esté en una lista se procede a borar los "**-**" *(2)* y por último, convertir la lista en un String con una coma y un espacio como delimitador *(3)*

```python
def elm_to_list(x):
    output=change.iloc[x].values.tolist() #(1)
    output[:] = (value for value in output if value != "-") #(2)
    output=", ".join(output) #(3)
    return output
```

- Se procede a iterar esta función sobre las filas y guardarla en una lista

```python
genre=[]
for i in change.index:
    genre.append(elm_to_list(i))
genre
```

- Se guarda transforma a una serie de pandas

```python
genre=pd.Series(genre)
genre
            
```

- Como el proceso está casi terminado antes de agregar la columna se borra primero las columnas que indican los géneros

```python
df_games.drop(list(change.columns), axis=1, inplace=True)
```

- Y por último, se agrega la columna género al dataframe

```python
df_games["genre"]=genre
df_games.head(3)
```
