### ¿Cómo empezar a usar Pyplot?
- Importamos Matplotlib y Pandas

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Creamos nuestras variables

```python
x=np.linspace(0,5,11)
y=x**2
```

- Graficar en función de y

```python
plt.plot(x,y,"yx-.")#se puede usar como segundo caracter, p, o, x, D entre otrosy como primero un color
plt.show()
```

```python
plt.plot(x,y,"gD-")
plt.show()
```

- Más info acerca de plot [aquí.](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot)
- Y para más gráficas [aquí.](https://matplotlib.org/stable/plot_types/basic/index.html)

#### Crear histograma


```python
plt.hist(y)
plt.show()
```

#### Crear gráfica de pie


```python
plt.pie(x)
plt.show()
```

#### Conocer correlación entre variables


```python
plt.scatter(x,y)  #correlacion entre la variable x e y
plt.show()
```

#### Distribución de los datos
- Distribución de los datos: Gráfico de caja


```python
plt.boxplot(y)
plt.show()
```
