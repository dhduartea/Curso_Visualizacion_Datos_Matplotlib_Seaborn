### **Subplot**
Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a través de índices:

#### Cómo acceder a la matriz de gráficos en Subplot

```python
plt.subplot(1,2,1)#(*args: Any, **kwargs: Any)
plt.plot(x,y)#Grafico 1

plt.subplot(1,2,2)
plt.hist(x)#Gráfico 2
plt.show()
```

- Ubicación Vertical 

```python
plt.subplot(2,1,1)#(filas, columnas, index)
plt.plot(x,y)

plt.subplot(2,1,2)# (dos filas, una columna, ubicado en la posición2)
plt.pie(x)
plt.show()
```

### Método orientado a objetos
Hay distintas maneras de hacer gráficas dentro de Matplotlib, ya se vio pyplot; sin embargo, no es tan sencillo personalizarla y entrar a ciertos parámetros.
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gráficas(axes), de las cuales, cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gráficas las cuales contienen diferentes ejes.

Es un poco más complicado, pero en el mismo gráfico podemos personalizarlo mucho mejor.

#### Diferencias
|      Pyplot     |           Object Oriented          |
|:---------------:|:----------------------------------:|
| Rápido          | Mayor personalización              |
| Fácil           | Más código                         |
| Una sola figura | Más amigable a múltiples diagramas |

#### Implementar el método orientado a objetos para graficar en Matplotlib

```python
x = np.arange(0, 10.5,0.5)
y = x ** 2
```

##### Parámetros de axes:
|   Pos. Eje x  |   Pos. Eje y  |   Size graph eje x    |   Size graph eje y|
|:-------------:|:-------------:|:---------------------:|:----------:|

- Utilizamos el concepto de objeto para crear una gráfica más personalizable

```python
fig= plt.figure()
axes=fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, "yx--")
fig.show()
```

```python
fig= plt.figure()
axes=fig.add_axes([0.5,20,0.4,0.5])
axes.plot(x,y, "y")
```

- Podemos crear varias gráficas dentro del lienzo llamando el método fig.

```python
fig= plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.8])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('yellow')
plt.show()
```

#### Generar una matriz de tipo fig, con dos filas y cuatro columnas

```python
fig, axes = plt.subplots(2, 4) # se pueden poner filas y columnas sin necesidad de poner los labels
axes[1,3].plot(x,y,'b')
axes[1,1].plot(y,x,'y')
plt.show()
```

- Alternativa para generar una matriz de tipo fig, accediendo a cada gráfica mediante los nombres dados en una tupla

```python


fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)
axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2, 'black')
axes5.hist(np.sin(x))
axes6.pie(x)
axes7.plot(y, np.sin(y), 'orange')
axes8.boxplot(y,'y')


#mejorar la visualización de las figuras
fig.tight_layout()
```

### Leyendas, etiquetas, títulos, tamaño
Para dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio.


#### Cómo dar contexto a los gráficos
- Creamos nuestras gráficas comunes

```python
x=np.arange(0,5,0.4)
y=np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, label="$sin(x)$")
axes[0].set_title("Relación x-y")#Título de gráfica 1
axes[0].set_xlabel("x")#Título eje x ax0
axes[0].set_ylabel("y")#Título eje y ax0
axes[0].legend() # para que aparezcala el label definido en plot

axes[1].plot(y,x, label="$sin(y)$")
axes[1].set_title("Relación y-x")
axes[1].set_xlabel("y")
axes[1].set_ylabel("x")
axes[1].legend(loc="lower right", #Loc pone el label en la posición deseada
        bbox_to_anchor=(0.5,0.2)) #bbbox define el ancho del label

```

#### Personalizar colores y estilos en Pyplot
Para personalizar con pyplot es un poco más largo cuando programamos, pero se puede personalizar muchísimo más



- Color


```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8') # Color RGB
ax.plot(x,x+2,color = '#66FF89')

ax.plot(x,x+3,color = 'blue') # Color común
ax.plot(x,x+4, color = 'white')
```

- Transparencia (alpha) y grosor (linewidth)


```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8', alpha= 0.5, linewidth= 18)
ax.plot(x,x+2,color = '#66FF89', linewidth= 3)
ax.plot(x,x+3,color = 'blue', linewidth= 5)
ax.plot(x,x+4, color = 'white', alpha = 0.3, linewidth= 12)
```

- Estilo (linestyle) y marcadores (marker)


```python
fig, ax = plt.subplots(figsize = (10,10))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = 'x')
ax.plot(x,x+4, color = 'green',linestyle = '--', marker = 'P')
```

- Tamaño de marcadores (markersize) y color (markerfacecolor)


```python
fig, ax = plt.subplots(figsize = (10,10))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = 'v', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4, color = 'green',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
```

- Para ver todos los estilos disponibles básicos

```python
for i in plt.style.available:
    plt.style.use(i)
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_title(i, size=20)
    ax.plot(x, y)
```

### Bar Plot
El gráfico de barras nos permite graficar variables categóricas, es decir, variables de texto, lo que es muy importante en el mundo de la ciencia de datos y Matplotlib ofrece ciertas características que nos facilita la vida en cuanto a graficar este tipo de variables.

- Variables categóricas


```python
countrys = ["CANADA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [38, 125, 129, 50, 83]
```

- Gráfica de barras


```python
plt.bar(countrys,population)
plt.show()
```

- Modificar colores y tamaño


```python
colors=["aqua","teal",  "green", "yellowgreen", "violet"]
plt.bar(countrys,population,  width=0.5, color= colors)
plt.show()
```

- Modificar los nombres de las variables en el eje ‘X’ con xticks


```python
plt.bar(countrys,population, width=0.5, color= colors)
plt.xticks(np.arange(5), ['Canada','Japon', 'Mexico', 'Colombia', 'Alemania'], rotation = 45)
plt.show()
```

- Crear un gráfico de manera horizontal con barh


```python
pop_median=np.median(population)
pop_mean=np.mean(population)
```

```python
plt.style.use("seaborn-darkgrid")
plot=plt.bar(countrys,population, color=colors)
plt.axhline(y = pop_mean, color = 'r', linestyle = '--', label='Population Mean')
plt.axhline(y = pop_median, color = 'orange', linestyle = '-.', label='Population Median')
plt.bar_label(plot, label_type='center', fontsize="12")
plt.ylabel('Población (mill)', fontsize="15")
plt.xlabel('Países', fontsize="15")
plt.title('POBLACIÓN POR PAÍS', fontsize="15", fontweight="20")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Seaborn es una librería construida sobre Matplotlib, por lo que hereda todas las bondades de la misma. Fue escrita por Michael Waskom y está integrada para estructuras de Pandas por lo que está optimizada para funcionar con DataFrames.

Seaborn tiene diferentes ventajas y entre ellas encuentras principalmente:
